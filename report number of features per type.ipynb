```python
import pandas as pd
```

```python
from src.ftr_aggregate import load_features
totals = ['NO_TOTAL', 'TOTAL']
years_choices = [('~2014', (2014, 2015, 2016, 2017)), ('~2016', (2016, 2017)), ]

features = {}
for total_or_not in totals:
    for year_type, years in years_choices:
        print(total_or_not, year_type)
        ftrs, cols_by_type = load_features(None, how=total_or_not, years=years, pair_with_y=False, verbose=True)
        features[(total_or_not, year_type)] = (ftrs, cols_by_type)
```

```python
def parse_output(total_year, sum_filtered, each_type):
    keep_lines = []
    for l in each_type.split('\n'):
        if l=='' or l.startswith('loading'):
            continue
        keep_lines.append(l.strip().replace('all columns #: ','').replace(' columns pass NA thres:', '').split(' '))

    print(total_year)
    print(sum_filtered)
    df = pd.DataFrame(keep_lines)
    df.columns = ['type', 'before', 'after']
    df.before = df.before.astype(int)
    df.after = df.after.astype(int)
    social = df.iloc[6:, 1:].sum().values
    roadnet = df.iloc[:6, 1:].sum().values
    print('%d(%d) road network features' % (roadnet[0], roadnet[1]))    
    print('%d(%d) social features' % (social[0], social[1]))
    dict_name = {
    'poi': 'POI', 'crash': 'crash data', '311': '311 requests', 'v0': 'vision zero', 
    'crime': 'crime data', 'moving': 'moving violations', 'parking': 'parking violations'
    }
    for name, before, after in df.iloc[6:].values:
        print('%d(%d) types for %s' %(before, after, dict_name[name]))
        
    return df
```
